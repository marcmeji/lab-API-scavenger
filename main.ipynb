### Challenge 2: Count Commits

Count how many commits were made in the past week.

1. Obtain all the commits made in the past week via API, which is a JSON array that contains multiple commit objects.

1. Count how many commit objects are contained in the array.

````python
flat_data = json_normalize(results)
flat_data.head()
````

````python
flat_data.columns
````

````python
labs_names = list(flat_data['name'])
print(labs_names)
````

````python
jsons_commits = []

for name in labs_names:
    url = 'https://api.github.com/repos/ta-datapt-bcn/' + name + '/commits'
    response= requests.get(url, auth=(username,tk))
    results = response.json()
    jsons_commits.append(results)
````

````python
list_commits = []
for commit in jsons_commits:
    for commit2 in commit:
        list_commits.append(commit2)
````
